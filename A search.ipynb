```python
import bisect
class Cola:
    def __init__(self):
        abstract

    def extend(self, items):
        for item in items: self.append(item)
        
    
class ColaPrioridad(Cola):
    """Cola ordenada respecto a un valor dado por una función f, que la clase
    tiene como atributo de dato. Admite orden creciente (min) y orden
    decreciente (max). Si el orden es min, el método pop devuelve y elimina el
    elemento de la cola con menor f(x). Si el orden es max, devuelve y elimina
    el de mayor f(x). En realidad, los elementos quie se guardan en la cola
    son tuplas (f(x),x). En Python, las tuplas se comparan lexicográficamente,
    así que se mantendrán ordenadas por su valor f(x). La pertenencia a la
    cola (x in C) se comprueba con x e ignorando f(x). Soporta además acceso
    como en diccionarios, ignorando igualmente el valor de f(x)"""


    def __init__(self, orden=min, f=lambda x: x):
        self.A=[]
        self.orden=orden 
        self.f=f

    def append(self, item): 
        bisect.insort(self.A, (self.f(item), item))

    def __len__(self):
        return len(self.A)

    def pop(self):
        if self.orden == min:
            return self.A.pop(0)[1]
        else:
            return self.A.pop()[1]

    def __contains__(self, item):
        for (_, x) in self.A:
            if x == item: return True
        return False

    def __getitem__(self, clave):
        for _, item in self.A:
            if item == clave:
                return item

    def __delitem__(self, clave):
        for i, (valor, item) in enumerate(self.A):
            if item == clave:
                self.A.pop(i)
                return

    def __str__(self):
        return str(self.A)
```

```python
import math
def distancia_manhattan_2(nodo, estado_fin):
    
    # ((xl_r - 1)**2)
    # ((x_r - xf_r)**2) + ((y_r - yf_r)**2)
    
    estado = nodo.estado
  
    xl_r = 0
    if (nodo.cargado):
        xl_r = 1
        
    x_r , y_r = estado[0][0] , estado[0][1]
    xf_r , xf_r = estado_fin[0][0] , estado_fin[0][1]
    x_m1 , y_m1 = estado[1][0] , estado[1][1]
    xf_m1 , yf_m1 = estado_fin[1][0] , estado_fin[1][1]
    x_m2 , y_m2 = estado[2][0] , estado[2][1]
    xf_m2 , yf_m2 = estado_fin[2][0] , estado_fin[2][1]
    x_m3 , y_m3 = estado[3][0] , estado[3][1]
    xf_m3 , yf_m3 = estado_fin[3][0] , estado_fin[3][1]
        
    hr = abs(xl_r - 1)
    hm1 = abs(x_m1 - xf_m1) + abs(y_m1 - yf_m1)
    hm2 = abs(x_m2 - xf_m2) + abs(y_m2 - yf_m2)
    hm3 = abs(x_m3 - xf_m3) + abs(y_m2 - yf_m3)
    
    h = (hr+1)*2 + hm1 + hm2 + hm3
    
    return h
```

```python
import math 
import networkx as nx
import matplotlib.pyplot as plt

class Nodo:
    def __init__(self, estado, parent, cargado, paquete):
        self.estado = estado
        self.parent = parent
        self.cargado = cargado
        self.paquete = paquete
        self.accion = " "
        self.H = 0
        self.G = 0
        self.F = self.G + self.H
    def __eq__(self, other):
        return self.estado == other.estado and self.cargado == other.cargado and self.paquete == other.paquete
    def __lt__(self, other):
        return self.F < other.F
    def __gt__(self, other):
        return self.F > other.F
  
def reconstruir_camino(nodo):
    camino = []
    while nodo.parent:
        camino.append(nodo)
        nodo = nodo.parent
    camino.append(nodo)
    print(len(camino))
    return camino[::-1]


def posicion_coincide(nodo):
    if (nodo.estado[0][0] == nodo.estado[1][0] and nodo.estado[0][1] == nodo.estado[1][1] ):
        return [1,True]
    elif (nodo.estado[0][0] == nodo.estado[2][0] and nodo.estado[0][1] == nodo.estado[2][1] ):
        return [2,True]
    elif (nodo.estado[0][0] == nodo.estado[3][0] and nodo.estado[0][1] == nodo.estado[3][1] ):
        return [3,True]
    return [0, False]
    
def paquete_en_la_meta(estado_actual, nodo_objetivo, indice):
    if (indice!=0):
        if( nodo_objetivo.estado[indice][0] == estado_actual[indice][0]
            and nodo_objetivo.estado[indice][1] == estado_actual[indice][1]
            and estado_actual[0][0]  ==  estado_actual[indice][0] 
            and estado_actual[0][1]  ==  estado_actual[indice][1]): 
                return True
    return False
    
    
def printar_bonito(nodo):
     print("\n"
           + " "+ " ACCION: "+str(nodo.accion)+ "\n POSICIONES: "
          " R: "+ str(nodo.estado[0]) + " P1: " + str(nodo.estado[1])+
          " P2: "+ str(nodo.estado[2]) +  " P3: "+ str(nodo.estado[3])
          )
        
def expandir(nodo_actual, nodo_objetivo, entorno):

    sucesores = []

    if (not nodo_actual.cargado and posicion_coincide(nodo_actual)[1] 
        and not paquete_en_la_meta(nodo_actual.estado, nodo_objetivo, posicion_coincide(nodo_actual)[0] )):
        
        sucesor = Nodo(nodo_actual.estado, nodo_actual, nodo_actual.cargado, nodo_actual.paquete)
        sucesor.paquete = posicion_coincide(nodo_actual)[0]
        sucesor.accion = "Cargar paquete "  + str(sucesor.paquete)
        sucesor.cargado = True
        sucesor.H = distancia_manhattan_2(sucesor, nodo_objetivo.estado)
        sucesor.G = nodo_actual.G+1
        sucesor.F = sucesor.H + sucesor.G
        sucesores.append(sucesor)
      
    elif nodo_actual.cargado and nodo_actual.paquete!=0 and paquete_en_la_meta(nodo_actual.estado, nodo_objetivo, nodo_actual.paquete):
        sucesor = Nodo(nodo_actual.estado, nodo_actual, False, 0)
        sucesor.accion = "Descargar paquete " + str(nodo_actual.paquete)
        sucesor.H = distancia_manhattan_2(sucesor, nodo_objetivo.estado)
        sucesor.G = nodo_actual.G+1
        sucesor.F = sucesor.H + sucesor.G
        sucesores.append(sucesor)
    
    else:      
        for posicion in [(0, -1), (0, 1), (-1, 0), (1, 0)]: 
            
            estado_actual = nodo_actual.estado
            posicion_actual_robot = estado_actual[0]
        
            nueva_posicion_robot = (posicion_actual_robot[0] + posicion[0], posicion_actual_robot[1] + posicion[1])

            if nueva_posicion_robot[0] > (len(entorno) - 1) or nueva_posicion_robot[0] < 0 or nueva_posicion_robot[1] > (len(entorno)-1) or nueva_posicion_robot[1] < 0:
                continue

            if entorno[nueva_posicion_robot[0]][nueva_posicion_robot[1]] == "#":
                continue
                
            estado_sucesor = [nueva_posicion_robot, estado_actual[1], estado_actual[2], estado_actual[3]]
            
            sucesor = Nodo(estado_sucesor, nodo_actual, False,0)
            sucesor.accion = "Mover Robot"
            
            if  nodo_actual.cargado and not paquete_en_la_meta(estado_sucesor, nodo_objetivo, nodo_actual.paquete) and nodo_actual.paquete!=0:
                estado_sucesor[nodo_actual.paquete] = nueva_posicion_robot
                sucesor.cargado = True
                sucesor.paquete = nodo_actual.paquete
                sucesor.estado = estado_sucesor
                sucesor.accion = str("Mover robot y paquete "+ str(sucesor.paquete))
                
            sucesor.H = distancia_manhattan_2(sucesor, nodo_objetivo.estado)
            sucesor.G = nodo_actual.G+1
            sucesor.F = sucesor.H + sucesor.G
            sucesores.append(sucesor)
    
    return sucesores
```
